# Backwards Euler's Method

### Importing necessary libraries.

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

### Initialising constants.

```python
S0 = 20 #Initial Susceptible People
I0 = 15 #Initial Infected People
R0 = 10 #Initial Released People
n = S0 + I0 + R0 # Population
h = 0.1 # Step Size for Time
beta = 0.01 # Infection Rate
gamma = 0.02 # Recovery Rate
```

### Initialising variable names for SymPy.

```python
s = sp.Symbol('s')  # the S1 t+1
i = sp.Symbol('i')  # the I t=1
r = sp.Symbol('r')  # the R t+1
```

### Approximating SIR using Backwards Euler's Method.

```python
S_list = [S0]
I_list = [I0]
R_list = [R0]
j = 0

for j in range(1, 200):
    n = S_list[j - 1] + I_list[j - 1] + R_list[j - 1]
    eq1 = sp.Eq(S_list[j - 1] - h * (beta * s * i), s)
    eq2 = sp.Eq(I_list[j - 1] + h * ((beta * s * i) - gamma * i), i)
    eq3 = sp.Eq(R_list[j - 1] + h * gamma * i, r)
    solution = sp.solve((eq1, eq2, eq3), (s, i, r))
    S_i, I_i, R_i = solution[1]
    S_list.append(S_i.evalf())
    I_list.append(I_i.evalf())
    R_list.append(R_i.evalf())
    j += 1
```

### Plotting the Results.

```python
XX = np.linspace(1, 20, 200)
plt.style.use('grayscale')
plt.plot(XX, S_list, label='Susceptible')
plt.plot(XX, I_list, label='Infected')
plt.plot(XX, R_list, label='Released')
plt.xlabel('Time')
plt.ylabel('Cases')
plt.legend()
plt.show()
```
