# Euler's Method

### Importing Necessary Libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Initialising constants.

```python
N = 45 #Population	
I0 = 15 # Initial number of infected people
R0 = 10 # Initial number of recovered people
S0 = N - I0 - R0 # Initial number of susceptible people
beta = 0.01 # Initial Infection Rate
gamma = 0.02 # Initial Recovery Rate
h = 1 # Step Size for Time
t = np.arange(0, 20 + h, h) # Time Range
```

### Creating Arrays for Each Variable.

```python
s = np.zeros(len(t)) 
s[0] = S0

i = np.zeros(len(t))
i[0] = I0

r = np.zeros(len(t))
r[0] = R0

y = [s, i, r]
```

### Defining Function for Partial Derivative Calculation,

```python
def deriv(y, t, beta, gamma):
    S = y[0]
    I = y[1]
    dSdt = -beta * S[t] * I[t]
    dIdt = beta * S[t] * I[t] - gamma * I[t]
    dRdt = gamma * I[t]
    return dSdt, dIdt, dRdt
```

### Approximating SIR Using Euler's Method

```python
for i in range(0, len(t) - 1):
    der = deriv(y, i, beta, gamma)
    for k in range(0, 3):
        y[k][i + 1] = y[k][i] + h * der[k]
```

### Plotting the Results.

```python
plt.style.use('grayscale')
plt.plot(t, y[0], label='Susceptible')
plt.plot(t, y[1], label='Infected')
plt.plot(t, y[2], label='Released')
plt.legend()
plt.title('Approximation of SIR Using Euler\'s Method')
plt.xlabel('Time')
plt.ylabel('Cases')
```
